```python
# Creating and training predictive model
model = NearestNeighbors(algorithm = "brute")
model.fit(movies_sparse)
```

APPLYING CLUSTERING WITH KMEANS

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Normalizing for clustering
scaler = StandardScaler()
movies_scaled = scaler.fit_transform(movie_pivot)
```

```python
# Applying KMeans
n_clusters = 8
k_means = KMeans(n_clusters=n_clusters, random_state = 42)
movie_pivot['Cluster'] = k_means.fit_predict(movies_scaled)
```

```python
display(movie_pivot)
```

```python
# Function to print clusters
def print_clusters(df):
  for cluster in range(n_clusters):
    print(f"\nCluster {cluster}:")
    movies_on_cluster = df[df['Cluster'] == cluster].index.tolist()
    for movie in movies_on_cluster:
      print(f" - {movie}")
```

```python
# Printing clusters
print_clusters(movie_pivot)
```

```python
# Recommendation function based on KMeans
def recommend_movie(movie, df):
  # finding input movie's cluster
  cluster_movie = df[df.index == movie]['Cluster'].values[0]

  # filtering movies of same cluster
  recommendations = df[df['Cluster'] == cluster_movie].index.tolist()
  recommendations.remove(movie)

  print(f"Filme de entrada: {movie} pertence ao Cluster {cluster_movie}")
  print(f"RecomendaÃ§Ãµes para quem gostou de {movie}: ")
  for movie in recommendations:
      print(f" - {movie}")
```

```python
recommend_movie('Ghost Rider', movie_pivot)
```

USING EUCLIDEAN DISTANCE

```python
def euclidean_distance(vetor1, vetor2):
  return np.sqrt(np.sum((vetor1 - vetor2) ** 2))
```

```python
# Recommendation function with KMeans and Euclidean Distance
def recommend_films_kmean_euclidean(film_input, df, n_recommendations=5):
  # finding movies' cluster
  cluster_film = df[df.index == film_input]['Cluster'].values[0]

  # filtering movies of same cluster
  recommendations = df[df['Cluster'] == cluster_film]
  recommendations = recommendations.drop(film_input, axis=0)

  # list to store the calculated distances
  distances = []

  # Input movie's vector
  vector_film_input = df.loc[film_input].drop(['Cluster']).values

  # calculate the euclidean distance between input movie and the other movies
  for filme in recommendations.index:
    vector_film = recommendations.loc[filme].drop(['Cluster']).values
    distance = euclidean_distance(vector_film_input, vector_film)
    distances.append((filme, distance))

  # sorting distances
  sorted_distances = sorted(distances, key=lambda x: x[1])

  print(f"filme de entrada: {film_input} pertence ao cluster {cluster_film}")
  print(f"Se vocÃ gostou de {film_input}, vocÃª tambÃ©m irÃ¡ gostar de:")

  for i in range(n_recommendations):
    recommended_film, dist = sorted_distances[i]
    print(f" - {recommended_film} (Euclidean distance: {dist:.2f})")
```

```python
# Testing recommendation system
film_test = 'Sunshine'
recommend_films_kmean_euclidean(film_test, movie_pivot)
```

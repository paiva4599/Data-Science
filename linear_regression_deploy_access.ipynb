```python
# Importando o arquivo de teste
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
```

```python
test = pd.read_csv("/content/drive/My Drive/Dados/casas_teste.csv")
```

```python
test = test.drop("Unnamed: 0", axis=1)
```

```python
# Fazendo o load do modelo
from joblib import load
model = load("regressao.joblib")
```

```python
# Exibindo o intercept
model.intercept_
```

```python
# Exibindo o coef
model.coef_
```

```python
# Fazendo a previsão
X_test = test.drop("MedHouseVal", axis=1)
```

```python
reg_pred = model.predict(X_test)
print(reg_pred)
```

```python
# Exportando esse arquivo
X_test["previsao"] = reg_pred
X_test.to_excel("previsoes_modelo.xlsx")
```
